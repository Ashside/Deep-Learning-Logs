```python
import torch
from torch import nn
from RNN import rnn

batch_size , num_steps = 32,35
train_iter,vocab = rnn.load_data_time_machine(batch_size,num_steps)
vocab_size = len(vocab)
num_hiddens = 256
num_layers = 2
num_inputs = vocab_size
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
lstm_layer = nn.LSTM(input_size=num_inputs, hidden_size=num_hiddens, num_layers=num_layers).to(device)
model = rnn.RNNModel(lstm_layer, vocab_size).to(device)
num_epochs = 500
lr = 2
rnn.train(model,train_iter=train_iter,vocab=vocab,lr=lr,num_epochs=num_epochs,device=device)
```
